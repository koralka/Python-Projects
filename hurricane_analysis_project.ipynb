```python
# damages (USD($)) of hurricanes
damages = ['Damages not recorded', '100M', 'Damages not recorded', '40M',
          '27.9M', '5M', 'Damages not recorded', '306M', '2M', '65.8M',
          '326M', '60.3M', '208M', '1.42B', '25.4M', 'Damages not recorded',
          '1.54B', '1.24B', '7.1B', '10B', '26.5B', '6.2B', '5.37B', '23.3B',
          '1.01B', '125B', '12B', '29.4B', '1.76B', '720M', '15.1B', '64.8B',
          '91.6B', '25.1B']

# 1
# Update Recorded Damages
conversion = {"M": 1000000,
             "B": 1000000000}

def update_damages_fun(damages):
    updated_damages = []
    for value in damages:
        if 'M' in value:
            updated_damages.append(float(value.replace('M', '')) * conversion ['M'])
        elif 'B' in value:
            updated_damages.append(float(value.replace('B', '')) * conversion ['B'])
        else:
            updated_damages.append(value)
    return updated_damages

updated_damages = update_damages_fun(damages)
print(updated_damages)
```

```python
# 3
# Organizing by Year
def organise_by_year(hurricanes):
    hurricanes_by_year = {}
    for hurricane in hurricanes:
        year = hurricanes[hurricane]['Year'] #to access year
        if year not in hurricanes_by_year:
            hurricanes_by_year[year] = [] #necessary to store multiple hurricanes that occured in the same year
        hurricanes_by_year[year].append(hurricanes[hurricane])
    return hurricanes_by_year

# create a new dictionary of hurricanes with year and key
hurricanes_by_year = organise_by_year(hurricanes)
pprint(hurricanes_by_year)
```

```python
# 4
# Counting Damaged Areas
def count_damaged_areas(hurricanes):
    frequency_of_area = {}
    for hurricane in hurricanes:
        affected_areas = hurricanes[hurricane]['Areas Affected']
        for area in affected_areas:
            if area not in frequency_of_area:
                frequency_of_area[area] = 1
            else:
                frequency_of_area[area] += 1
    return frequency_of_area
        
# create dictionary of areas to store the number of hurricanes involved in
hurricanes_areas_count = count_damaged_areas(hurricanes)
pprint(hurricanes_areas_count)
```

```python
# 5
# Calculating Maximum Hurricane Count
def calculate_max_count(hurricanes):
    max_hurricane_count = sum(list(hurricanes.values()))
    return max_hurricane_count

total_hurricanes = calculate_max_count(hurricanes_areas_count)
print(f"The total amount of hurricanes is {total_hurricanes}")

# find most frequently affected area and the number of hurricanes involved in
def find_area_affected_most(hurricanes):
    max_area_name = max(hurricanes, key=hurricanes.get)
    max_area_count = hurricanes[max_area_name]

    return max_area_name, max_area_count
    
area_affected_most = find_area_affected_most(hurricanes_areas_count)
print(f"The area affected by hurricanes the most is {area_affected_most}.")
```

```python
# 6
# Calculating the Deadliest Hurricane
def find_deadliest_hurricane(hurricanes):
    max_hurricane = None
    max_deaths = 0
    for hurricane, data in hurricanes.items():
        deaths = data['Deaths']
        if deaths > max_deaths:
            max_hurricane = hurricane
            max_deaths = deaths
    return max_hurricane, max_deaths

hurricane, deaths = find_deadliest_hurricane(hurricanes)
# find highest mortality hurricane and the number of deaths
print(f"The hurricane that caused the most deaths is {hurricane} with {deaths} deaths.")
```

```python
# 7
# Rating Hurricanes by Mortality
# categorize hurricanes in new dictionary with mortality severity as key
mortality_scale = {0: 0,
1: 100,
2: 500,
3: 1000,
4: 10000}

def rate_by_mortality(hurricanes):
  mortality_rate = {0: [], 1: [], 2: [], 3: [], 4: [], 5: []} 
  for hurricane in hurricanes:
    mortality_count = hurricanes[hurricane]['Deaths']
    if mortality_count <= 0:
      mortality_rate[0].append(hurricane),
    elif mortality_count > 0 and mortality_count <=100:
      mortality_rate[1].append(hurricane),
    elif mortality_count > 100 and mortality_count <= 500:
      mortality_rate[2].append(hurricane),
    elif mortality_count > 200 and mortality_count <= 10000:
      mortality_rate[3].append(hurricane),
    elif mortality_count > 300 and mortality_count <= 10000:
      mortality_rate.append(hurricane),
    else:
      mortality_rate[5].append(hurricane)
  return mortality_rate

hurricanes_by_mortality = rate_by_mortality(hurricanes)
pprint(hurricanes_by_mortality)
```

```python
# 8
# Calculating Hurricane Maximum Damage
import copy
def convert_damage(hurricanes):
  conversion = {'M': 1000000, 'B': 1000000000}
  hurricanes_dmg_updated = copy.deepcopy(hurricanes)
  for hurricane, data in hurricanes_dmg_updated.items():
    dmg = data['Damage']
    if isinstance(dmg, str):
      if dmg == 'Damages not recorded':
        continue
      if 'M' in dmg:
        hurricanes_dmg_updated[hurricane]['Damage'] = float(dmg.replace('M', '')) * conversion['M']
      elif 'B' in dmg:
        hurricanes_dmg_updated[hurricane]['Damage'] = float(dmg.replace('B', '')) * conversion['B']
  return hurricanes_dmg_updated

new_hurricanes = convert_damage(hurricanes)
pprint(new_hurricanes)

# find highest damage inducing hurricane and its total cost
def get_max_damage(new_hurricanes):
  max_damage = 0
  max_hurricane = None 
  for hurricane, data in new_hurricanes.items():
    dmg = data['Damage']
    if isinstance(dmg, float):
      if dmg > max_damage:
        max_damage = dmg
        max_hurricane = hurricane
  return max_damage, max_hurricane
    
max_damage, max_hurricane = get_max_damage(new_hurricanes)
print(f"The hurricane that caused the most damage is {max_hurricane} with {max_damage} dollars worth of damage.")
```

```python
# 9
# Rating Hurricanes by Damage
# categorize hurricanes in new dictionary with damage severity as key
damage_scale = {0: 0,
                1: 100000000,
                2: 1000000000,
                3: 10000000000,
                4: 50000000000}

def rate_by_damage(hurricanes):
  damage_rate = {0: [], 1: [], 2: [], 3: [], 4: []}
  for hurricane in hurricanes:
    damage_count = new_hurricanes[hurricane]['Damage']
    if isinstance (damage_count, str):
      continue
    if damage_count <= 0:
      damage_rate[0].append(hurricane),
    elif damage_count >= 0 and damage_count <= damage_scale[1]:
      damage_rate[1].append(hurricane),
    elif damage_count >= damage_scale[1] and damage_count <= damage_scale[2]:
      damage_rate[2].append(hurricane),
    elif damage_count >= damage_scale[2] and damage_count <= damage_scale[3]:
      damage_rate[3].append(hurricane),
    else:
      damage_rate[4].append(hurricane)
  return damage_rate

hurricanes_by_damage = rate_by_damage(new_hurricanes)
pprint(hurricanes_by_damage)
```
